## Problem 2

Consider the following code to encode a text string as a binary string using ASCII.  As we did in class, we use a string of seven bits for each character.

````python
def encode(text):
    encoded_text = str()
    for letter in text:
        encoded_text = encoded_text + '{0:07b}'.format(ord(letter))
    return encoded_text

def decode(text):
    decoded_text = str()
    while len(text) > 0:
        temp = text[:7]
        text = text[7:]
        decoded_text = decoded_text + chr(int(temp,2))
    return decoded_text
````

Let's try it out:

````python
text = 'I love spaghetti and meat loaf.'
encoded_text = encode(text)
print(encoded_text)

decoded_text = decode(encoded_text)
print(decoded_text)
````

Since the ASCII code corresponding to a character is a number between 0 and 127, we encode each character using seven bits.  If the ASCII code in binary has less than seven bits, then we pad it to the left with zeros.

The following code can be used to xor two different strings of bits of the same length.  (You can modify this code if you'd like to handle two strings of different lengths...but be careful with paddings in this case...)

````python
# xor two bits
def bit_xor(bit1,bit2):
    if bit1 == bit2:
        return '0'
    else:
        return '1'
    
# xor two strings of bits with the same length
def xor(A,B):
    
    if len(A) != len(B):
        print("The strings don't have the same number of characters.")
        return 0
    
    result = str()
    for x in range(len(A)):
            result = result + bit_xor(A[x],B[x])  
    
    return result
````
